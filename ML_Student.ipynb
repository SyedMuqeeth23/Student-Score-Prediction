```python
import numpy as np
import pandas as pd
import sklearn
```

```python
df = pd.read_csv("/Users/muqeeth/Desktop/ML_Student 1_Project/Student Scores Data (1).csv")
```

```python
print(df.isnull().sum())
```

```python
df.dropna(inplace=True)
```

```python
print(df.isnull().sum())
```

```python
df=df[df['Attendance (%)'].str.isnumeric()]
```

```python
df
```

```python
df=df[df['Weekly_Test_Score'].str.isnumeric()]
```

```python
df=df[df['Final_Exam_Score'].str.isnumeric()]
```

```python
df
```

```python
df.info()
```

```python
df['Attendance (%)']=df['Attendance (%)'].astype(int)
```

```python
df.info()
```

```python
df['Weekly_Test_Score']=df['Weekly_Test_Score'].astype(int)
```

```python
df['Final_Exam_Score']=df['Final_Exam_Score'].astype(int)
```

```python
df.info()
```

```python
from sklearn.model_selection import train_test_split

df.info()
```

```python
x = df[["Study_Hours","Attendance (%)","Weekly_Test_Score"]]
y = df[["Final_Exam_Score"]]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=47)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import r2_score
from sklearn.metrics import root_mean_squared_error
from sklearn.metrics import mean_absolute_error

r2 = r2_score(y_pred,y_test)
print(r2)

rmse = root_mean_squared_error(y_pred,y_test)
print(rmse)

mae = mean_absolute_error(y_pred,y_test)
print(mae)
```

```python
import joblib

joblib.dump(model , "lr.pkl")
```
